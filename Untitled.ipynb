```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
file_path = r"C:\Users\leube\Ironhack\Bootcamp\week_5\amz_uk_price_prediction_dataset.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# number 1.1
frequency_table = df['category'].value_counts()
print(frequency_table)
```

```python
top_5_categories = frequency_table.head()
print(f'The top 5 categories are: {top_5_categories}')
```

```python
#number 1.2
plt.figure(figsize=(12, 6))
top_5_categories.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_5_categories.value_counts().plot.pie(autopct='%1.1f%%', startangle=90,  colors=sns.color_palette("Set3"))
```

```python
# number 2.1
print(df['price'].mean())
print(df['price'].median())
print(df['price'].mode())
# the average price is a lot higher than the mode
```

```python
# number 2.2
var = df['price'].var()
std = df['price'].std()
range = df['price'].max() - df['price'].min()
IQR = df['price'].quantile([0.25, 0.5])

print(var)
print(std)
print(range)
print(IQR)

# the product prices vary on average by 345 so there is a significant variation in the prices
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['price'], vert=False)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# most products fall into the lower price spectrum and there are some extreme outliers
```

```python
# number 3.1
print(df['stars'].mean())
print(df['stars'].median())
print(df['stars'].mode())

# customers generally rate the products badly
```

```python
# number 3.2
var = df['stars'].var()
std = df['stars'].std()
range = df['stars'].max() - df['stars'].min()
IQR = df['stars'].quantile([0.25, 0.5])

print(var)
print(std)
print(range)
print(IQR)

# there is significant variation on customer feedback
```

```python
skewness_price = df['stars'].skew()
kurtosis_price = df['stars'].kurtosis()

skewness_price, kurtosis_price
```
